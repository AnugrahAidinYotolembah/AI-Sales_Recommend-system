# E. PREPROCESSING

```python
# Install pycaret terlebih dahulu jika belum diinstall
# pip install pycaret

import pandas as pd
from pycaret.nlp import *

# Memanggil dataset yang sudah ada sebelumnya
df = pd.read_csv('AI_SALES FITUR 6.csv')  # Gantilah 'your_dataset.csv' dengan nama dataset sesuai kebutuhan

# Membuat objek NLP dari pycaret
nlp = setup(data=df, target='produk_training_name', session_id=123, log_experiment=True, experiment_name='nlp_experiment')

# Menampilkan hasil preprocessing
preprocess_results = create_model('lda')

# Menampilkan model yang dihasilkan
print(preprocess_results)

# Menggabungkan semua data untuk mengevaluasi model
evaluate_model(preprocess_results)

# Melakukan prediksi menggunakan model yang telah dibuat
predict_model(preprocess_results)

# Finalizing model
final_nlp_model = finalize_model(preprocess_results)

# Menampilkan hasil akhir model
print(final_nlp_model)

# Save the model
save_model(final_nlp_model, 'final_nlp_model')

# Load the model
loaded_model = load_model('final_nlp_model')

# Contoh penggunaan model yang telah disimpan
new_data = pd.DataFrame({'produk_training_name': ['Text yang ingin diprediksi']})
prediction = predict_model(loaded_model, data=new_data)
print(prediction)
```
